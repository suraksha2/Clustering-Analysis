Clustering Analysis

```python
# Load Dataset
import pandas as pd

file_path = 'EastWestAirlines.xlsx'
df = pd.read_excel(file_path, sheet_name=1)  
df.head()
```

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
```

```python
# Load dataset
df = pd.read_excel('EastWestAirlines.xlsx')
df.head()
```

```python

# Convert to numeric safely
df = df.apply(pd.to_numeric, errors='coerce')

# Fill missing values (DO NOT drop rows)
df = df.fillna(df.mean())

# Drop columns that are completely NaN
df = df.dropna(axis=1, how='all')

# Drop ID column if exists
df = df.drop(columns=['ID'], errors='ignore')

print("Shape after cleaning:", df.shape)
df.head()
```

```python
#SCALING

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

print("Any NaNs:", scaled_df.isna().sum().sum())
scaled_df.head()
```

```python
#EDA

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
 
df.hist(figsize=(12,10))
plt.show()
```

```python
#ELBOW METHOD

inertia = []
K = range(1, 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia, 'bo-')
plt.xlabel('Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.show()
```

```python
#KMEANS 

kmeans = KMeans(n_clusters=3, random_state=42)
labels_k = kmeans.fit_predict(scaled_df)

scaled_df['KMeans'] = labels_k

print("Silhouette Score:", silhouette_score(scaled_df.drop('KMeans', axis=1), labels_k))
```

```python
#DBSCAN

db = DBSCAN(eps=1.5, min_samples=5)
labels_d = db.fit_predict(scaled_df.drop('KMeans', axis=1))

scaled_df['DBSCAN'] = labels_d

if len(set(labels_d)) > 1:
    print("Silhouette Score:", silhouette_score(scaled_df.drop(['KMeans','DBSCAN'], axis=1), labels_d))
else:
    print("DBSCAN formed one cluster - tune eps")
```

```python
#Visualization

plt.scatter(scaled_df.iloc[:,0], scaled_df.iloc[:,1], c=scaled_df['KMeans'])
plt.title("KMeans Clusters")
plt.show()

plt.scatter(scaled_df.iloc[:,0], scaled_df.iloc[:,1], c=scaled_df['DBSCAN'])
plt.title("DBSCAN Clusters")
plt.show()
```

Insights
- No rows were dropped → data preserved
- Missing values handled using mean
- KMeans works well for structured clusters
- DBSCAN detects noise but needs tuning


```python
# DBSCAN Experimentation with different eps and min_samples
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import numpy as np

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.select_dtypes(include=[np.number]))

eps_values = [0.5, 1.0, 1.5, 2.0]
min_samples_values = [3, 5, 10]

results = []

for eps in eps_values:
    for min_samples in min_samples_values:
        db = DBSCAN(eps=eps, min_samples=min_samples)
        labels = db.fit_predict(X_scaled)

        # Ignore cases where only one cluster is formed
        if len(set(labels)) > 1 and -1 not in set(labels):
            score = silhouette_score(X_scaled, labels)
        else:
            score = -1

        results.append((eps, min_samples, score))

# Display results
import pandas as pd
results_df = pd.DataFrame(results, columns=['eps', 'min_samples', 'silhouette_score'])
results_df.sort_values(by='silhouette_score', ascending=False)
```
